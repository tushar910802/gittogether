```python
def scrape_product_details(url):
    # Set up the WebDriver (make sure to specify the path to your WebDriver if it's not in your PATH)
    driver = webdriver.Chrome()

    # Open the webpage
    driver.get(url)

    # Wait for the page to load
    driver.implicitly_wait(10)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')

    # Find all product containers
    products = soup.find_all('div', class_='product-card')

    # List to store product details
    product_list = []

    # Loop through each product and extract details
    for product in products:
        product_name = product.find('a', class_='product-card__name').text.strip()
        product_price = product.find('span', class_='price-item price-item--regular').text.strip()
        product_url = product.find('a', class_='product-card__name')['href']
        product_url = 'https://www.westside.com' + product_url

        # Open the product page to get the description
        driver.get(product_url)
        time.sleep(3)  # Wait for the page to load
        product_page_source = driver.page_source
        product_soup = BeautifulSoup(product_page_source, 'html.parser')
        product_description = product_soup.find('div', class_='product-single__description').text.strip()

        product_list.append({
            'Name': product_name,
            'Price': product_price,
            'URL': product_url,
            'Description': product_description
        })

    # Close the WebDriver
    driver.quit()

    return product_list


def find_best_price(product_list, product_name):
    df = pd.DataFrame(product_list)
    filtered_df = df[df['Name'].str.contains(product_name, case=False, na=False)]
    if not filtered_df.empty:
        best_price = filtered_df['Price'].min()
        best_offer = filtered_df.loc[filtered_df['Price'] == best_price]
        return best_offer.to_dict('records')
    else:
        return None
```

```python
# URL of the webpage to scrape
url = 'https://www.westside.com/collections/perfumes-and-fragrances'
product_name = input("Enter the product name: ")

# Scrape product details
product_list = scrape_product_details(url)

# Find the best price and promo offers
best_offer = find_best_price(product_list, product_name)

if best_offer:
    print("Best Offer:")
    for offer in best_offer:
        print(f"Name: {offer['Name']}")
        print(f"Price: {offer['Price']}")
        print(f"URL: {offer['URL']}")
        print(f"Description: {offer['Description']}")
        print('-' * 20)
else:
    print("No offers found for the given product name.")
```
